```python
def crop_yield(region, weights):
    result = 0
    for x, w in zip(region, weights):
        result += x * w
    return result
```

```python
import numpy as np

# Creating the example array
arr = np.arange(1, 10)
print("Original array:", arr)
```

```python
# Creating arrays filled with zeros and ones
zeros = np.zeros(9)
ones = np.ones(9)
linspace_arr = np.linspace(1, 9, 9)

print("Array of zeros:", zeros)
print("Array of ones:", ones)
print("Linearly spaced array:", linspace_arr)
```

```python
linspace_arr = np.linspace(1, 9, 10001)
print("Linearly spaced array:", linspace_arr)
```

```python
# Reshape the array to 3x3
reshaped_arr = arr.reshape(3, 3)
print("Reshaped array (3x3):", reshaped_arr)

# Flatten the reshaped array
flattened_arr = reshaped_arr.flatten()
print("Flattened array:", flattened_arr)

# Concatenate arrays
concatenated_arr = np.concatenate((arr, arr))
print("Concatenated array:", concatenated_arr)

# Stack arrays vertically and horizontally
vstacked_arr = np.vstack((arr, arr,arr))
hstacked_arr = np.hstack((arr, arr,arr))
print("Vertically stacked array:\n", vstacked_arr)
print("Horizontally stacked array:", hstacked_arr)

# Split array into 3 sub-arrays
split_arr = np.split(arr, 1)
print("Split array into 3 sub-arrays:", split_arr)
```

```python
# Element-wise addition, subtraction, multiplication, division
add_arr = np.add(arr, 2)
sub_arr = np.subtract(arr, 2)
mul_arr = np.multiply(arr, 2)
div_arr = np.divide(arr, 2)

print("Addition:", add_arr)
print("Subtraction:", sub_arr)
print("Multiplication:", mul_arr)
print("Division:", div_arr)

# Dot product (with itself reshaped as a 3x3 matrix)
dot_product = np.dot(reshaped_arr, reshaped_arr)
print("Dot product:\n", dot_product)

# Square root, exponential, logarithm
sqrt_arr = np.sqrt(arr)
exp_arr = np.exp(arr)
log_arr = np.log(arr)

print("Square root:", sqrt_arr)
print("Exponential:", exp_arr)
print("Logarithm:", log_arr)

# Sum, mean, standard deviation, variance, min, max
sum_arr = np.sum(arr)
mean_arr = np.mean(arr)
std_arr = np.std(arr)
var_arr = np.var(arr)
min_arr = np.min(arr)
max_arr = np.max(arr)

print("Sum:", sum_arr)
print("Mean:", mean_arr)
print("Standard deviation:", std_arr)
print("Variance:", var_arr)
print("Min:", min_arr)
print("Max:", max_arr)
```

```python
# Median, percentile
median_arr = np.median(arr)
percentile_arr = np.percentile(arr, 50)

print("Median:", median_arr)
print("50th percentile:", percentile_arr)

# Correlation coefficient matrix, covariance matrix
corrcoef_arr = np.corrcoef(reshaped_arr)
cov_arr = np.cov(reshaped_arr)

print("Correlation coefficient matrix:\n", corrcoef_arr)
print("Covariance matrix:\n", cov_arr)
```

```python
# Inverse, determinant, eigenvalues and eigenvectors
# For these operations, we need a square matrix, so we use reshaped_arr (3x3)
inverse_arr = np.linalg.inv(reshaped_arr)
det_arr = np.linalg.det(reshaped_arr)
eigvals, eigvecs = np.linalg.eig(reshaped_arr)
u, s, vh = np.linalg.svd(reshaped_arr)
sol = np.linalg.solve(reshaped_arr, [1, 2, 3])

print("Inverse of matrix:\n", inverse_arr)
print("Determinant of matrix:", det_arr)
print("Eigenvalues:", eigvals)
print("Eigenvectors:\n", eigvecs)
print("Singular Value Decomposition (U, S, V^H):\n", u, s, vh)
print("Solution of linear matrix equation:", sol)
```

```python
import numpy as np

# Creating arrays
empty_arr = np.empty((3, 3))
eye_arr = np.eye(3)
full_arr = np.full((3, 3), 7)
print("Empty array:\n", empty_arr)
print("Identity matrix:\n", eye_arr)
print("Full array with sevens:\n", full_arr)

# Array attributes and information
arr = np.array([1, 2, 3, 4, 5, 6]).reshape(2, 3)
print("Array shape:", arr.shape)
print("Array size:", arr.size)
print("Array dtype:", arr.dtype)

# Sorting and searching
sorted_arr = np.sort(arr, axis=1)
indices_sorted = np.argsort(arr, axis=1)
print("Sorted array:\n", sorted_arr)
print("Indices that would sort the array:\n", indices_sorted)

# Set operations
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])
unique_arr = np.unique(arr1)
intersect_arr = np.intersect1d(arr1, arr2)
union_arr = np.union1d(arr1, arr2)
setdiff_arr = np.setdiff1d(arr1, arr2)
print("Unique elements:", unique_arr)
print("Intersection:", intersect_arr)
print("Union:", union_arr)
print("Set difference:", setdiff_arr)

# Broadcasting and other miscellaneous operations
a = np.array([1, 2, 3])
b = np.array([[4], [5], [6]])
broadcasted_sum = a + b
print("Broadcasted sum:\n", broadcasted_sum)

# Transpose and flattening
transposed_arr = np.transpose(arr)
raveled_arr = np.ravel(arr)
print("Transposed array:\n", transposed_arr)
print("Flattened array:", raveled_arr)

# Clipping values
clipped_arr = np.clip(arr, 2, 5)
print("Clipped array:", clipped_arr)
```
